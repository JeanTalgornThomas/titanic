```python
import pandas as pd
import matplotlib.pyplot as mp
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
%matplotlib inline
```

```python
# Chargement CSV Train
train = pd.read_csv("data/train.csv")
train.tail()
```

```python
#Trie des données du train
train = train[train['Age'].notna()]
train = train[train['Sex'].notna()]
train = train[train['Pclass'].notna()]
train = train.drop(columns=['SibSp', 'Parch', 'Fare', 'Ticket' ,'Cabin', 'Embarked'])
train.tail()
```

```python
# Chargement CSV Test
test = pd.read_csv("data/test.csv")
test.tail()
```

```python
#Trie des données du train
test = test[test['Age'].notna()]
test = test[test['Sex'].notna()]
test = test[test['Pclass'].notna()]
test = test.drop(columns=['SibSp', 'Parch', 'Fare', 'Ticket' ,'Cabin', 'Embarked'])
test.tail()
```

```python
# Création du Decision Tree CLassifier
clf = DecisionTreeClassifier()

clf = clf.fit(train)

#Predict the response for test dataset
y_pred = clf.predict(test)
```
